```python
import numpy as np
import pandas as pd
import anndata as ad
from scipy.sparse import csr_matrix
from scipy.io import mmwrite
from scipy import sparse
import os
import sys
print(ad.__version__)
```

```python
bmmc = ad.read_h5ad("data/raw_data/rna_atac/BMMC/GSE194122_openproblems_neurips2021_multiome_BMMC_processed.h5ad")
```

```python
bmmc
```

```python
# Making directories and saving files by batch
output_path = "data/processed_data/rna_atac/"
file_paths = {}
for batch in bmmc.obs["batch"].unique():
    # making directory for each batch
    full_path = os.path.join(output_path, "BMMC-"+batch)
    print(full_path)
    # file_paths[directory] = full_path
    os.makedirs(full_path, exist_ok=True)
    # saving each batch data
    bmmc_batch = bmmc[bmmc.obs["batch"] == batch]
    # split into unimodal data
    rna = bmmc_batch[:, bmmc_batch.var["feature_types"]=="GEX"]
    rna
    atac = bmmc_batch[:, bmmc_batch.var["feature_types"]=="ATAC"]
    atac
    # RNA counts
    mmwrite(os.path.join(full_path, "rna_counts.mtx"), rna.layers["counts"].transpose())
    # ATAC counts
    mmwrite(os.path.join(full_path, "atac_counts.mtx"), atac.layers["counts"].transpose())
    # ACTIVITY
    mmwrite(os.path.join(full_path, "activity_counts.mtx"),atac.obsm["ATAC_gene_activity"].transpose())
    # barcode
    rna.obs_names.to_series().to_csv(os.path.join(full_path, "barcode.csv"), header=["x"], index=False)
    # cell type
    rna.obs["cell_type"].to_csv(os.path.join(full_path, "ct_annotation.csv"), header=["x"], index=False)
    # genes for RNA counts matrix
    rna.var_names.to_series().to_csv(os.path.join(full_path, "genes.csv"), header=["x"], index=False)
    # peaks for ATAC counts matrix
    atac.var_names.to_series().to_csv(os.path.join(full_path, "peaks.csv"), header=["x"], index=False)
    # genes for activity matrix
    df = pd.DataFrame(atac.uns["ATAC_gene_activity_var_names"], columns=['x'])
    df.to_csv(os.path.join(full_path, "activity_genes.csv"), index=False)
```
